# Project 2: Feature Engineering and Optimization Analysis
## Phases 1, 2 & 3: Data Loading, Collinearity, and Feature Extraction


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.datasets import load_breast_cancer, load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, FastICA, TruncatedSVD
from IPython.display import display

# --- Create results directory ---
results_dir = 'results'
os.makedirs(results_dir, exist_ok=True)
print(f"Directory '{results_dir}' ensured.")

# ===================================================================
# PHASE 1: DATA LOADING & EXPLORATORY DATA ANALYSIS (EDA)
# ===================================================================

# -------------------------------------------------------------------
# Classification Data (Wisconsin Breast Cancer)
# -------------------------------------------------------------------
print("\n--- 1.1: Loading & Exploring Classification Data (Breast Cancer) ---")
data_cancer = load_breast_cancer()
X_cancer = pd.DataFrame(data_cancer.data, columns=data_cancer.feature_names)
y_cancer = pd.Series(data_cancer.target, name='target')
print(f"Cancer Shape (Samples, Features): {X_cancer.shape}")
print("\nFirst 5 rows of features:")
display(X_cancer.head())
print("\nData types and missing values:")
X_cancer.info()
print("\nDescriptive statistics:")
display(X_cancer.describe())
print("\nTarget variable distribution (0=Malignant, 1=Benign):")
print(y_cancer.value_counts())

# -------------------------------------------------------------------
# Regression Data (Boston Housing)
# -------------------------------------------------------------------
print("\n\n--- 1.2: Loading & Exploring Regression Data (Boston Housing) ---")
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
feature_names_boston = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
X_boston = pd.DataFrame(data, columns=feature_names_boston)
y_boston = pd.Series(target, name='MEDV')
print(f"Boston Shape (Samples, Features): {X_boston.shape}")
print("\nFirst 5 rows of features:")
display(X_boston.head())
print("\nData types and missing values:")
X_boston.info()
print("\nDescriptive statistics:")
display(X_boston.describe())
print("\nTarget variable distribution (Median Value of Homes):")
sns.histplot(y_boston, kde=True)
plt.title('Distribution of Boston Housing Prices (MEDV)')
plt.xlabel('Median Value ($1000s)')
plt.ylabel('Frequency')
plt.show()

# -------------------------------------------------------------------
# Clustering Data (UCI Iris)
# -------------------------------------------------------------------
print("\n\n--- 1.3: Loading & Exploring Clustering Data (Iris) ---")
data_iris = load_iris()
X_iris = pd.DataFrame(data_iris.data, columns=data_iris.feature_names)
y_iris_true = pd.Series(data_iris.target, name='true_species') # Ground truth for later evaluation
print(f"Iris Shape (Samples, Features): {X_iris.shape}")
print("\nFirst 5 rows of features:")
display(X_iris.head())
print("\nData types and missing values:")
X_iris.info()
print("\nDescriptive statistics:")
display(X_iris.describe())
print("\nTrue species distribution (for later evaluation):")
print(y_iris_true.value_counts().rename({0: 'Setosa', 1: 'Versicolour', 2: 'Virginica'}))


print("\n\nPhase 1 (Data Loading & EDA) completed successfully.")
```

```python
# ===================================================================
# PHASE 2: INITIAL COLLINEARITY ANALYSIS
# ===================================================================
print("\n--- Phase 2: Analyzing Initial Feature Collinearity ---")
sns.set_theme(style="white")

# Breast Cancer ---
print("\nAnalyzing Breast Cancer dataset...")
corr_cancer = X_cancer.corr()
mask = np.triu(np.ones_like(corr_cancer, dtype=bool))
f, ax = plt.subplots(figsize=(18, 15))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr_cancer, mask=mask, cmap=cmap, vmax=1.0, vmin=-1.0, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Correlation Matrix of Breast Cancer Features', fontsize=16)
plt.tight_layout()
save_path = os.path.join(results_dir, 'breast_cancer_corr.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.show()
print(f"Breast Cancer correlation heatmap saved.")

# Boston Housing ---
print("\nAnalyzing Boston Housing dataset...")
corr_boston = X_boston.corr()
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr_boston, cmap=cmap, annot=True, fmt=".2f", vmax=1.0, vmin=-1.0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Correlation Matrix of Boston Housing Features', fontsize=16)
plt.tight_layout()
save_path = os.path.join(results_dir, 'boston_housing_corr.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.show()
print(f"Boston Housing correlation heatmap saved.")

#  Iris ---
print("\nAnalyzing Iris dataset...")
corr_iris = X_iris.corr()
f, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr_iris, cmap=cmap, annot=True, fmt=".2f", vmax=1.0, vmin=-1.0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Correlation Matrix of Iris Features', fontsize=16)
plt.tight_layout()
save_path = os.path.join(results_dir, 'iris_corr.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.show()
print(f"Iris correlation heatmap saved.")

print("\n\nPhase 2 (Collinearity Analysis) completed successfully.")
```

```python
from sklearn.model_selection import train_test_split

print("\n--- Phase 2.5: Creating Train-Validation-Test Splits ---")
print("This is the most critical step to prevent data leakage.")


TEST_SIZE = 0.2  # 20% of the data for the final test set
VALIDATION_SIZE = 0.25 # 25% of the REMAINING data (i.e., 20% of original) for validation


data_splits = {}


X_cancer_train_full, X_cancer_test, y_cancer_train_full, y_cancer_test = train_test_split(
    X_cancer, y_cancer, test_size=TEST_SIZE, random_state=42, stratify=y_cancer
)

X_cancer_train, X_cancer_val, y_cancer_train, y_cancer_val = train_test_split(
    X_cancer_train_full, y_cancer_train_full, test_size=VALIDATION_SIZE, random_state=42, stratify=y_cancer_train_full
)
data_splits['Cancer'] = {
    'X_train': X_cancer_train, 'y_train': y_cancer_train,
    'X_val': X_cancer_val, 'y_val': y_cancer_val,
    'X_test': X_cancer_test, 'y_test': y_cancer_test
}
print(f"Cancer splits -> Train: {X_cancer_train.shape}, Validation: {X_cancer_val.shape}, Test: {X_cancer_test.shape}")


X_boston_train_full, X_boston_test, y_boston_train_full, y_boston_test = train_test_split(
    X_boston, y_boston, test_size=TEST_SIZE, random_state=42
)
X_boston_train, X_boston_val, y_boston_train, y_boston_val = train_test_split(
    X_boston_train_full, y_boston_train_full, test_size=VALIDATION_SIZE, random_state=42
)
data_splits['Boston'] = {
    'X_train': X_boston_train, 'y_train': y_boston_train,
    'X_val': X_boston_val, 'y_val': y_boston_val,
    'X_test': X_boston_test, 'y_test': y_boston_test
}
print(f"Boston splits -> Train: {X_boston_train.shape}, Validation: {X_boston_val.shape}, Test: {X_boston_test.shape}")


X_iris_train, X_iris_test, y_iris_train, y_iris_test = train_test_split(
    X_iris, y_iris_true, test_size=TEST_SIZE, random_state=42, stratify=y_iris_true
)
data_splits['Iris'] = {
    'X_train': X_iris_train, 'y_train': y_iris_train,
    'X_test': X_iris_test, 'y_test': y_iris_test,
    'X_full': X_iris, 'y_full': y_iris_true # Keep full data for clustering
}
print(f"Iris splits   -> Train: {X_iris_train.shape}, Test: {X_iris_test.shape}, Full: {X_iris.shape}")

print("\nData splitting complete. Feature engineering will now be performed correctly on these partitions.")

```

```python
# ===================================================================
# PHASE 3: FEATURE EXTRACTION 
# ===================================================================
print("\n--- Phase 3: Feature Extraction based on Linear Independence ---")
print("Refactored to fit scalers and transformers ONLY on the training set.")

print("\n--- 3.1: Standardizing Data (Fitting only on Train sets) ---")

scalers = {}
scaled_data = {}

for name, splits in data_splits.items():
    if 'X_train' in splits: 
        scaler = StandardScaler()

        scaled_train = pd.DataFrame(scaler.fit_transform(splits['X_train']), columns=splits['X_train'].columns)
        
        scaled_val = pd.DataFrame(scaler.transform(splits['X_val']), columns=splits['X_val'].columns) if 'X_val' in splits else None
        scaled_test = pd.DataFrame(scaler.transform(splits['X_test']), columns=splits['X_test'].columns)

        scalers[name] = scaler
        scaled_data[name] = {
            'train': scaled_train,
            'val': scaled_val,
            'test': scaled_test
        }
        print(f"'{name}' dataset standardized.")

iris_scaler = StandardScaler()
scaled_data['Iris']['full'] = pd.DataFrame(iris_scaler.fit_transform(data_splits['Iris']['X_full']), columns=data_splits['Iris']['X_full'].columns)
scalers['Iris_full'] = iris_scaler
print("'Iris' full dataset also standardized for clustering.")


print("\n--- 3.2: PCA vs SVD Explained Variance Analysis (Demonstrating Equivalence) ---")

datasets_for_analysis = {
    "Cancer": scaled_data['Cancer']['train'],
    "Boston": scaled_data['Boston']['train'],
    "Iris": scaled_data['Iris']['full'] # For clustering, we analyze variance on the full set
}
n_components_95 = {}


fig, axes = plt.subplots(1, 3, figsize=(24, 6))
fig.suptitle('Cumulative Explained Variance: PCA vs. SVD on Standardized Data', fontsize=16, y=1.02)

for i, (name, data) in enumerate(datasets_for_analysis.items()):
    ax = axes[i]
    
    pca = PCA(n_components=None, random_state=42)
    pca.fit(data)
    cum_var_pca = np.cumsum(pca.explained_variance_ratio_)
    
    n_features = data.shape[1]
    svd = TruncatedSVD(n_components=n_features - 1, random_state=42)
    svd.fit(data)
    cum_var_svd = np.cumsum(svd.explained_variance_ratio_)


    n_comp = np.argmax(cum_var_pca >= 0.95) + 1
    n_components_95[name] = n_comp

    ax.plot(range(1, len(cum_var_pca) + 1), cum_var_pca, 'o-', color='blue', lw=2, label='PCA Variance')

    ax.plot(range(1, len(cum_var_svd) + 1), cum_var_svd, 'x--', color='red', lw=2, label='SVD Variance')
    ax.axhline(y=0.95, color='r', linestyle='-', label='95% Threshold')
    ax.axvline(x=n_comp, color='k', linestyle=':', label='Optimal Components')
    
    ax.set_title(f'Dataset: {name}')
    ax.set_xlabel('Number of Components'); ax.set_ylabel('Cumulative Variance Ratio')
    ax.grid(True, which='both', linestyle='--'); ax.legend()

plt.tight_layout(rect=[0, 0, 1, 0.96])
save_path = os.path.join(results_dir, 'pca_svd_explained_variance_comparison.png')
plt.savefig(save_path, dpi=300)
plt.show()

print("Optimal number of components to retain 95% variance (based on PCA on train/full data):")
for name, n_comp in n_components_95.items():
    print(f"- {name}: {n_comp} components")


print("\n--- 3.3: Transforming Data with PCA, TruncatedSVD, and FastICA ---")

transformed_data = {} 

for name, splits in scaled_data.items():
    if 'train' not in splits: continue # Skip if no train set
    
    n_comp = n_components_95[name]
    transformed_data[name] = {'Original_Scaled': splits} # Store the scaled original data

    transformers = {
        'PCA': PCA(n_components=n_comp, random_state=42),
        'SVD': TruncatedSVD(n_components=n_comp, random_state=42),
        'ICA': FastICA(n_components=n_comp, random_state=42, max_iter=1000, tol=0.01)
    }
    
    for trans_name, trans_obj in transformers.items():
        trans_obj.fit(splits['train'])
        
        X_train_transformed = trans_obj.transform(splits['train'])
        X_val_transformed = trans_obj.transform(splits['val']) if splits['val'] is not None else None
        X_test_transformed = trans_obj.transform(splits['test'])
        
        transformed_data[name][trans_name] = {
            'train': X_train_transformed,
            'val': X_val_transformed,
            'test': X_test_transformed
        }

    if name == 'Iris':
        transformed_data[name]['Original_Scaled']['full'] = splits['full']
        for trans_name, trans_obj in transformers.items():
            trans_obj_full = transformers[trans_name] 
            trans_obj_full.fit(splits['full'])
            transformed_data[name][trans_name]['full'] = trans_obj_full.transform(splits['full'])

    print(f"'{name}' dataset transformed. New shape for train set (PCA/SVD/ICA): {transformed_data[name]['PCA']['train'].shape}")

print("\nPhase 3 (Feature Extraction) completed successfully, with no data leakage.")

```

```python
# ===================================================================
# PHASE 4: FEATURE SELECTION 
# ===================================================================
print("\n--- Phase 4: Applying Supervised Feature Selection Methods ---")
print("Refactored to fit selectors ONLY on the training set.")

from sklearn.feature_selection import SelectKBest, f_classif, f_regression, RFE
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from IPython.display import display


selected_data_final = {}
analysis_tables = {} # To store feature ranking tables

print("\n--- 4.1: Processing Breast Cancer (Classification) ---")
k_cancer = n_components_95['Cancer']
print(f"Target number of features: {k_cancer}")


cancer_selectors = {
    'SelectKBest': SelectKBest(f_classif, k=k_cancer),
    'RFE-Linear': RFE(estimator=LogisticRegression(max_iter=5000, random_state=42), n_features_to_select=k_cancer),
    'RFE-Tree': RFE(estimator=RandomForestClassifier(random_state=42), n_features_to_select=k_cancer)
}

selected_data_final['Cancer'] = {}
for name, selector in cancer_selectors.items():
    print(f"Fitting '{name}' selector on Cancer training data...")

    selector.fit(scaled_data['Cancer']['train'], data_splits['Cancer']['y_train'])
    
    X_train_selected = selector.transform(scaled_data['Cancer']['train'])
    X_val_selected = selector.transform(scaled_data['Cancer']['val'])
    X_test_selected = selector.transform(scaled_data['Cancer']['test'])
    
    selected_data_final['Cancer'][name] = {
        'train': X_train_selected,
        'val': X_val_selected,
        'test': X_test_selected
    }

selector_kbest_cancer = cancer_selectors['SelectKBest']
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(15, 8))
scores = pd.Series(selector_kbest_cancer.scores_, index=X_cancer.columns).sort_values(ascending=False)
bars = ax.bar(scores.index, scores.values, color=sns.color_palette("viridis", len(scores)))
ax.set_title('Feature Importance Scores (f_classif) for Breast Cancer (from Train Set)', fontsize=16, weight='bold')
ax.set_ylabel('ANOVA F-value'); ax.tick_params(axis='x', rotation=90, labelsize=10)
ax.bar_label(bars, fmt='%.1f', padding=3, fontsize=9)
plt.tight_layout()
plt.savefig(os.path.join(results_dir, 'cancer_f_scores_enhanced.png'), dpi=300)
plt.show()

# Analysis Table
cancer_analysis_df = pd.DataFrame({
    'Feature': X_cancer.columns,
    'SelectKBest_Score': cancer_selectors['SelectKBest'].scores_,
    'RFE_Linear_Rank': cancer_selectors['RFE-Linear'].ranking_,
    'RFE_Tree_Rank': cancer_selectors['RFE-Tree'].ranking_
}).sort_values(by='SelectKBest_Score', ascending=False).reset_index(drop=True)
analysis_tables['Cancer'] = cancer_analysis_df
print("\n--- Cancer Feature Selection Analysis Table (from Train Set) ---")
display(cancer_analysis_df.head(15))
cancer_analysis_df.to_csv(os.path.join(results_dir, 'cancer_features_analysis.csv'), index=False)
cancer_analysis_df.head(15).to_latex(os.path.join(results_dir, 'cancer_features_table.tex'), index=False, float_format="%.2f")


rank_df_cancer = pd.DataFrame({
    'SelectKBest': cancer_analysis_df['SelectKBest_Score'].rank(method='first', ascending=False),
    'RFE (Linear)': cancer_analysis_df['RFE_Linear_Rank'],
    'RFE (Tree)': cancer_analysis_df['RFE_Tree_Rank']
})
rank_df_cancer.index = cancer_analysis_df['Feature']
plt.figure(figsize=(8, 12))
sns.heatmap(rank_df_cancer.sort_values(by='SelectKBest'), annot=True, cmap="viridis_r", fmt=".0f", linewidths=.5, cbar_kws={'label': 'Feature Rank (Lower is Better)'})
plt.title('Feature Rank Comparison for Breast Cancer', fontsize=16, weight='bold')
plt.xlabel('Selection Method', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.tight_layout()
save_path = os.path.join(results_dir, 'cancer_feature_rank_comparison.png')
plt.savefig(save_path, dpi=300)
plt.show()


print("\n\n--- 4.2: Processing Boston Housing (Regression) ---")
k_boston = n_components_95['Boston']
print(f"Target number of features: {k_boston}")


boston_selectors = {
    'SelectKBest': SelectKBest(f_regression, k=k_boston),
    'RFE-Linear': RFE(estimator=LinearRegression(), n_features_to_select=k_boston),
    'RFE-Tree': RFE(estimator=RandomForestRegressor(random_state=42), n_features_to_select=k_boston)
}

selected_data_final['Boston'] = {}
for name, selector in boston_selectors.items():
    print(f"Fitting '{name}' selector on Boston training data...")
    selector.fit(scaled_data['Boston']['train'], data_splits['Boston']['y_train'])
    X_train_selected = selector.transform(scaled_data['Boston']['train'])
    X_val_selected = selector.transform(scaled_data['Boston']['val'])
    X_test_selected = selector.transform(scaled_data['Boston']['test'])
    selected_data_final['Boston'][name] = {'train': X_train_selected, 'val': X_val_selected, 'test': X_test_selected}

selector_kbest_boston = boston_selectors['SelectKBest']
fig, ax = plt.subplots(figsize=(12, 7))
scores = pd.Series(selector_kbest_boston.scores_, index=X_boston.columns).sort_values(ascending=False)
bars = ax.bar(scores.index, scores.values, color=sns.color_palette("plasma", len(scores)))
ax.set_title('Feature Importance Scores (f_regression) for Boston Housing (from Train Set)', fontsize=16, weight='bold')
ax.set_ylabel('F-value'); ax.tick_params(axis='x', rotation=45, labelsize=10)
ax.bar_label(bars, fmt='%.1f', padding=3, fontsize=9)
plt.tight_layout()
plt.savefig(os.path.join(results_dir, 'boston_f_scores_enhanced.png'), dpi=300)
plt.show()

boston_analysis_df = pd.DataFrame({
    'Feature': X_boston.columns,
    'SelectKBest_Score': boston_selectors['SelectKBest'].scores_,
    'RFE_Linear_Rank': boston_selectors['RFE-Linear'].ranking_,
    'RFE_Tree_Rank': boston_selectors['RFE-Tree'].ranking_
}).sort_values(by='SelectKBest_Score', ascending=False).reset_index(drop=True)
analysis_tables['Boston'] = boston_analysis_df
print("\n--- Boston Feature Selection Analysis Table (from Train Set) ---")
display(boston_analysis_df)
boston_analysis_df.to_csv(os.path.join(results_dir, 'boston_features_analysis.csv'), index=False)
boston_analysis_df.to_latex(os.path.join(results_dir, 'boston_features_table.tex'), index=False, float_format="%.2f")

rank_df_boston = pd.DataFrame({
    'SelectKBest': boston_analysis_df['SelectKBest_Score'].rank(method='first', ascending=False),
    'RFE (Linear)': boston_analysis_df['RFE_Linear_Rank'],
    'RFE (Tree)': boston_analysis_df['RFE_Tree_Rank']
})
rank_df_boston.index = boston_analysis_df['Feature']
plt.figure(figsize=(8, 8))
sns.heatmap(rank_df_boston.sort_values(by='SelectKBest'), annot=True, cmap="viridis_r", fmt=".0f", linewidths=.5, cbar_kws={'label': 'Feature Rank (Lower is Better)'})
plt.title('Feature Rank Comparison for Boston Housing', fontsize=16, weight='bold')
plt.xlabel('Selection Method', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.tight_layout()
save_path = os.path.join(results_dir, 'boston_feature_rank_comparison.png')
plt.savefig(save_path, dpi=300)
plt.show()


print("\n\nPhase 4 (Feature Selection) completed successfully, with no data leakage.")

```

```python
# ===================================================================
# PHASE 5a: REGRESSION ANALYSIS 
# ===================================================================
print("\n--- Phase 5a: Regression Analysis (Comprehensive & Methodologically Sound) ---")
print("Using train/validation/test splits correctly. No data leakage.")

# --- Imports for this cell ---
from sklearn.linear_model import LinearRegression, SGDRegressor, Ridge
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import time
import warnings
from IPython.display import display

from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)

boston_feature_sets = {
    'Original': transformed_data['Boston']['Original_Scaled'],
    'PCA': transformed_data['Boston']['PCA'],
    'ICA': transformed_data['Boston']['ICA'],
    'SVD': transformed_data['Boston']['SVD'],
    'SelectKBest': selected_data_final['Boston']['SelectKBest'],
    'RFE-Linear': selected_data_final['Boston']['RFE-Linear'],
    'RFE-Tree': selected_data_final['Boston']['RFE-Tree']
}

y_train = data_splits['Boston']['y_train']
y_val = data_splits['Boston']['y_val']
y_test = data_splits['Boston']['y_test']

regression_results = {}


print("\n... 5a.1: Running LinearRegression (Stability) on all 7 datasets ...")
coef_stability_data = []
for name, data_partitions in boston_feature_sets.items():
    model_lr = LinearRegression().fit(data_partitions['train'], y_train)
    y_pred = model_lr.predict(data_partitions['test'])
    mse = mean_squared_error(y_test, y_pred)
    coef_magnitude = np.linalg.norm(model_lr.coef_)
    
    result_entry = { 'Dataset': name, 'Model': 'LinearRegression', 'Test_MSE': mse, 'Coef_Magnitude': coef_magnitude }
    regression_results[f'LR_{name}'] = result_entry
    coef_stability_data.append(result_entry)
    print(f"[LinearRegression] - {name}: Test MSE = {mse:.4f}, Coef. Magnitude = {coef_magnitude:.4f}")

plt.figure(figsize=(12, 7)); sns.set_theme(style="whitegrid")
df_stability = pd.DataFrame(coef_stability_data)
ax = sns.barplot(data=df_stability, x='Dataset', y='Coef_Magnitude', palette='rocket')
ax.set_yscale("log")
plt.title('Coefficient Stability Analysis (Log Scale) - All Datasets', fontsize=16)
plt.ylabel('Coefficient Vector L2 Norm (Log Scale)'); plt.xlabel('Feature Set')
plt.xticks(rotation=30); plt.tight_layout()
plt.savefig(os.path.join(results_dir, 'regression_stability_plot_comprehensive.png'), dpi=300)
plt.show()


print("\n... 5a.2: Comparing Ridge (on Original) vs. LR (on PCA) ...")
lr_pca_result = regression_results['LR_PCA']
# Train Ridge on the 'Original_Scaled' training data
model_ridge = Ridge(alpha=1.0, random_state=42).fit(boston_feature_sets['Original']['train'], y_train)
y_pred_ridge = model_ridge.predict(boston_feature_sets['Original']['test'])
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
coef_mag_ridge = np.linalg.norm(model_ridge.coef_)

print(f"[LR on PCA]     - Test MSE = {lr_pca_result['Test_MSE']:.4f}, Coef. Magnitude = {lr_pca_result['Coef_Magnitude']:.4f}")
print(f"[Ridge on Orig] - Test MSE = {mse_ridge:.4f}, Coef. Magnitude = {coef_mag_ridge:.4f}")
regression_results['Ridge_Original'] = {'Dataset': 'Original', 'Model': 'Ridge(alpha=1.0)', 'Test_MSE': mse_ridge, 'Coef_Magnitude': coef_mag_ridge}


print("\n... 5a.3: Running SGDRegressor (Convergence on Train/Validation Sets) ...")

n_epochs = 100
convergence_data = []

for name, data_partitions in boston_feature_sets.items():
    print(f"Running SGD convergence test for: {name}")
    model_sgd = SGDRegressor(max_iter=1, warm_start=True, penalty=None, learning_rate='constant', eta0=0.01, random_state=42, tol=None)
    
    for epoch in range(n_epochs):
        model_sgd.fit(data_partitions['train'], y_train)

        train_mse = mean_squared_error(y_train, model_sgd.predict(data_partitions['train']))
        val_mse = mean_squared_error(y_val, model_sgd.predict(data_partitions['val']))
        convergence_data.append({'Dataset': name, 'Epoch': epoch + 1, 'MSE': train_mse, 'Set': 'Train'})
        convergence_data.append({'Dataset': name, 'Epoch': epoch + 1, 'MSE': val_mse, 'Set': 'Validation'})
        
    final_test_mse = mean_squared_error(y_test, model_sgd.predict(data_partitions['test']))
    regression_results[f'SGD_{name}'] = {'Dataset': name, 'Model': 'SGDRegressor', 'Test_MSE': final_test_mse, 'N_Iter': n_epochs}
    print(f"[SGDRegressor] - {name}: Final Test MSE = {final_test_mse:.4f}")

plt.figure(figsize=(14, 8))
ax = plt.gca()
df_convergence = pd.DataFrame(convergence_data)

sns.lineplot(data=df_convergence, x='Epoch', y='MSE', hue='Dataset', style='Set', 
             linewidth=2.5, markers=False, palette='Set1', ax=ax)

ax.set_title('SGDRegressor Convergence: Train vs. Validation MSE (Enhanced Clarity)', fontsize=16)
ax.set_xlabel('Epoch'); ax.set_ylabel('Mean Squared Error')
ax.legend(title='Feature Set / Data Split')

ax.set_ylim(bottom=20, top=30) 
ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1)) 
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.5)) 

ax.grid(True, which='both', linestyle='--')
plt.tight_layout()
plt.savefig(os.path.join(results_dir, 'regression_convergence_plot_train_val.png'), dpi=300)
plt.show()


df_regression_results = pd.DataFrame.from_dict(regression_results, orient='index')
print("\n--- Final Regression Results Summary (Corrected) ---")
display(df_regression_results)
df_regression_results.to_csv(os.path.join(results_dir, 'regression_results_summary_comprehensive.csv'))
print("Comprehensive regression summary table saved.")

```

```python
# ===================================================================
# PHASE 5b: CLUSTERING ANALYSIS 
# ===================================================================
print("--- Stage 5b: Clustering Analysis (KMeans) ---")

# --- Imports for this cell ---
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score
import time


true_k = 3

kmeans_params = {'n_clusters': true_k, 'n_init': 'auto', 'random_state': 42}


y_iris_full = data_splits['Iris']['y_full']
datasets_to_cluster = {
    'Original (4D)': transformed_data['Iris']['Original_Scaled']['full'],
    'PCA (2D)':      transformed_data['Iris']['PCA']['full'],
    'ICA (2D)':      transformed_data['Iris']['ICA']['full'],
    'SVD (2D)':      transformed_data['Iris']['SVD']['full']
}

clustering_results = []
visual_data = {} # To store data for plotting


print("Running KMeans on all Iris feature sets...")
for name, data in datasets_to_cluster.items():
    start_time = time.time()
    model_kmeans = KMeans(**kmeans_params).fit(data)
    comp_time = time.time() - start_time
    
    clustering_results.append({
        'Feature Set': name,
        'Inertia': model_kmeans.inertia_,
        'Silhouette': silhouette_score(data, model_kmeans.labels_),
        'ARS': adjusted_rand_score(y_iris_full, model_kmeans.labels_),
        'Iterations': model_kmeans.n_iter_,
        'Time (s)': comp_time
    })
    visual_data[name] = model_kmeans.labels_
    
df_clustering_results = pd.DataFrame(clustering_results).set_index('Feature Set')
print("\n--- Clustering Performance Analysis Table ---")
display(df_clustering_results)
df_clustering_results.to_csv(os.path.join(results_dir, 'clustering_results_summary.csv'))

print("\nGenerating comprehensive visualization plot...")

X_2D = transformed_data['Iris']['PCA']['full']


plot_df = pd.DataFrame(X_2D, columns=['PCA Component 1', 'PCA Component 2'])
plot_df['True Labels'] = y_iris_full.map({0: 'Setosa', 1: 'Versicolour', 2: 'Virginica'})
plot_df['KMeans (Original 4D)'] = visual_data['Original (4D)']
plot_df['KMeans (PCA 2D)'] = visual_data['PCA (2D)']
plot_df['KMeans (ICA 2D)'] = visual_data['ICA (2D)']

fig, axes = plt.subplots(2, 2, figsize=(16, 12), sharex=True, sharey=True)
sns.set_theme(style="whitegrid")
fig.suptitle('KMeans Clustering Analysis on Iris Dataset (k=3)', fontsize=20)

ax = axes[0, 0]
sns.scatterplot(data=plot_df, x='PCA Component 1', y='PCA Component 2', hue='True Labels', palette='deep', ax=ax, s=60)
ax.set_title('Ground Truth (True Species)', fontsize=16)

plot_map = { 'KMeans (PCA 2D)': axes[0, 1], 'KMeans (Original 4D)': axes[1, 0], 'KMeans (ICA 2D)': axes[1, 1] }
for hue, ax in plot_map.items():
    sns.scatterplot(data=plot_df, x='PCA Component 1', y='PCA Component 2', hue=hue, palette='deep', ax=ax, s=60, legend=False)
    ax.set_title(f'KMeans trained on {hue.split(" ")[2]} data', fontsize=16)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig(os.path.join(results_dir, 'clustering_visualization.png'), dpi=300)
plt.show()
```

```python
# ===================================================================
# PHASE 5c: CLASSIFICATION ANALYSIS 
# ===================================================================
print("\n--- Phase 5c: Classification Analysis (KNN vs. RF) ---")
print("Using train/validation/test splits correctly. No data leakage.")

# --- Imports for this cell ---
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import time

cancer_feature_sets = {
    'Original': transformed_data['Cancer']['Original_Scaled'],
    'PCA': transformed_data['Cancer']['PCA'],
    'ICA': transformed_data['Cancer']['ICA'],
    'SVD': transformed_data['Cancer']['SVD'],
    'SelectKBest': selected_data_final['Cancer']['SelectKBest'],
    'RFE-Linear': selected_data_final['Cancer']['RFE-Linear'],
    'RFE-Tree': selected_data_final['Cancer']['RFE-Tree']
}
y_train_cancer = data_splits['Cancer']['y_train']
y_val_cancer = data_splits['Cancer']['y_val']
y_test_cancer = data_splits['Cancer']['y_test']

models_to_test = {
    'KNN (k=5)': KNeighborsClassifier(n_neighbors=5),
    'RandomForest': RandomForestClassifier(random_state=42)
}
classification_results = []

print("\nRunning comprehensive classification analysis...")
for model_name, model_obj in models_to_test.items():
    print(f"--- Testing Model: {model_name} ---")
    for name, data_partitions in cancer_feature_sets.items():
        X_train, X_test = data_partitions['train'], data_partitions['test']
        
        fit_start = time.time()
        model_obj.fit(X_train, y_train_cancer)
        fit_time = time.time() - fit_start
        
        predict_start = time.time()
        y_pred = model_obj.predict(X_test)
        predict_time = time.time() - predict_start
        
        accuracy = accuracy_score(y_test_cancer, y_pred)
        
        classification_results.append({
            'Model': model_name, 'Feature Set': name, 'Accuracy': accuracy,
            'Fit Time (s)': fit_time, 'Predict Time (s)': predict_time
        })

df_classification_results = pd.DataFrame(classification_results)
df_pivot = df_classification_results.pivot(index='Feature Set', columns='Model', values=['Accuracy', 'Fit Time (s)', 'Predict Time (s)'])
print("\n--- Comprehensive Classification Performance Table ---")
display(df_pivot)
df_pivot.to_csv(os.path.join(results_dir, 'classification_results_summary.csv'))

fig, axes = plt.subplots(1, 2, figsize=(18, 7)); sns.set_theme(style="whitegrid")
sns.barplot(data=df_classification_results, x='Accuracy', y='Feature Set', hue='Model', ax=axes[0], palette='viridis')
axes[0].set_title('Model Accuracy Comparison', fontsize=16); axes[0].set_xlabel('Test Set Accuracy'); axes[0].set_xlim(0.9, 1.0)
sns.barplot(data=df_classification_results, x='Predict Time (s)', y='Feature Set', hue='Model', ax=axes[1], palette='viridis')
axes[1].set_title('Predict Time Comparison (Log Scale)', fontsize=16); axes[1].set_xlabel('Predict Time (s) - Log Scale'); axes[1].set_xscale('log')
plt.suptitle('Classification Performance: KNN vs. RandomForest', fontsize=20); plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig(os.path.join(results_dir, 'classification_accuracy_predict_time.png'), dpi=300); plt.show()

print("\n... Analyzing KNN k-sensitivity using the validation set ...")
k_range = list(range(1, 31))
knn_sensitivity_results = []
datasets_for_k_test = { 'Original': cancer_feature_sets['Original'], 'PCA': cancer_feature_sets['PCA'] }

for name, data_partitions in datasets_for_k_test.items():
    X_train, X_val = data_partitions['train'], data_partitions['val']
    for k in k_range:
        model_k = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train_cancer)
        val_accuracy = model_k.score(X_val, y_val_cancer)
        knn_sensitivity_results.append({'Dataset': name, 'k': k, 'Validation Accuracy': val_accuracy})

df_k_sensitivity = pd.DataFrame(knn_sensitivity_results)
best_k_orig = df_k_sensitivity.loc[df_k_sensitivity[df_k_sensitivity['Dataset'] == 'Original']['Validation Accuracy'].idxmax()]
best_k_pca = df_k_sensitivity.loc[df_k_sensitivity[df_k_sensitivity['Dataset'] == 'PCA']['Validation Accuracy'].idxmax()]
print(f"Best k for Original data (based on val set): {int(best_k_orig['k'])} with accuracy {best_k_orig['Validation Accuracy']:.4f}")
print(f"Best k for PCA data (based on val set): {int(best_k_pca['k'])} with accuracy {best_k_pca['Validation Accuracy']:.4f}")

plt.figure(figsize=(12, 7))
sns.lineplot(data=df_k_sensitivity, x='k', y='Validation Accuracy', hue='Dataset', style='Dataset', linewidth=2.5, markers=True)
plt.title('KNN Performance vs. k (Evaluated on Validation Set)', fontsize=16)
plt.xlabel('Number of Neighbors (k)'); plt.ylabel('Validation Set Accuracy')
plt.grid(True, which='both', linestyle='--'); plt.tight_layout()
plt.savefig(os.path.join(results_dir, 'classification_knn_k_sensitivity.png'), dpi=300); plt.show()
```
